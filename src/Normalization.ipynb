```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
#Đọc file train sau khi preprocessing
train_new_data = pd.read_csv('../Data/train_cleaned.csv')

#Áp dụng Min Max Scaling cho các biến số trừ cột "id"
scaler = MinMaxScaler()
num_col = [col for col in train_new_data.columns if train_new_data[col].dtype in ['int64', 'float64'] and col != 'id']
train_new_data[num_col] = scaler.fit_transform(train_new_data[num_col])

if 'feature_12' in train_new_data.columns:

    # Bỏ cột feature_12 gốc
    train_new_data = train_new_data.drop(columns=['feature_12'])

# Xác định các cột categorical (các cột kiểu object)
categorical_cols = train_new_data.select_dtypes(include=['object']).columns.tolist()

# Mã hóa các cột categorical bằng LabelEncoder
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    train_new_data[col] = le.fit_transform(train_new_data[col].astype(str))
    label_encoders[col] = le

# Lưu lại file train đã chuẩn hóa
train_new_data.to_csv('../Data/train_normalized.csv', index=False)

# In ra 5 dòng đầu tiên của file train sau khi chuẩn hóa
train_new_data.head()
```

```python
#Đọc file test sau khi preprocessing
test_new_data = pd.read_csv('../Data/test_cleaned.csv')

#Áp dụng Min Max Scaling cho các biến số trừ cột "id"
scaler = MinMaxScaler()
num_col = [col for col in test_new_data.columns if test_new_data[col].dtype in ['int64', 'float64'] and col != 'id']
test_new_data[num_col] = scaler.fit_transform(test_new_data[num_col])

if 'feature_12' in test_new_data.columns:

    # Bỏ cột feature_12 gốc
    test_new_data = test_new_data.drop(columns=['feature_12'])

# Xác định các cột categorical (các cột kiểu object)
categorical_cols = test_new_data.select_dtypes(include=['object']).columns.tolist()

# Mã hóa các cột categorical bằng LabelEncoder
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    test_new_data[col] = le.fit_transform(test_new_data[col].astype(str))
    label_encoders[col] = le

#Lưu lại file test đã được chuẩn hóa vào file mới 
test_new_data.to_csv('../Data/test_normalized.csv', index=False)

#In ra 5 dòng đầu tiên của file test sau khi chuẩn hóa 
test_new_data.head()
```

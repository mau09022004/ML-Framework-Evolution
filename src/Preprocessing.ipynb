```python
import pandas as pd
```

```python
#Đọc dữ liệu train
train_data = pd.read_csv("../Data/train.csv")
train_data.info()
train_data.describe()
train_data.head()
```

```python
#Đọc dữ liệu test
test_data = pd.read_csv("../Data/test.csv")
test_data.info()
test_data.describe()
test_data.head()
```

```python
# Xử lý giá trị thiếu và trùng lặp trực tiếp trên train_data
# Điền giá trị thiếu cho các cột kiểu số (int64, float64)
for col in train_data.select_dtypes(include=['int64', 'float64']).columns:
    train_data[col] = train_data[col].fillna(train_data[col].mean())  # Gán lại giá trị sau khi điền

# Điền giá trị thiếu cho các cột kiểu chuỗi (object) bằng 'Non'
for col in train_data.select_dtypes(include=['object']).columns:
    train_data[col] = train_data[col].fillna('Non')  # Gán lại giá trị sau khi điền

# Loại bỏ dòng trùng lặp
train_data.drop_duplicates(inplace=True)

# Lưu dữ liệu đã xử lý vào một file mới
train_data.to_csv("../Data/train_cleaned.csv", index=False)
```

```python
# Xử lý giá trị thiếu và trùng lặp trực tiếp trên test_data
# Điền giá trị thiếu cho các cột kiểu số (int64, float64)
for col in test_data.select_dtypes(include=['int64', 'float64']).columns:
    test_data[col] = test_data[col].fillna(test_data[col].mean())  # Gán lại giá trị sau khi điền

# Điền giá trị thiếu cho các cột kiểu chuỗi (object) bằng 'Non'
for col in test_data.select_dtypes(include=['object']).columns:
    test_data[col] = test_data[col].fillna('Non')  # Gán lại giá trị sau khi điền

# Loại bỏ dòng trùng lặp
test_data.drop_duplicates(inplace=True)

# Lưu dữ liệu đã xử lý vào một file mới
test_data.to_csv("../Data/test_cleaned.csv", index=False)
```
